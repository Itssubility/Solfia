## preprocessing 함수
- preprocessing_QA(`keywords`, `filename`)
    - `keywords`: 원하는 성격 키워드 list로 입력
    - `filename`: json 파일명(example.json)
- preprocessing_persona(`target`, `filename`)
    - `target`: 원하는 단일 말투
    - `filename`: json 파일명(example.json)
    - `persona_df` return됨
- preprocessing_all(`personality`, `qa_file`, `persona_df`, `filename`)
    - `personality`: persona 정보
    - `qa_file`: QA file 경로
    - `persona_df`: preprocessing_persona()에서 return된 persona_df
    - `filename`: json 파일명(example.json)

```python
def preprocessing_QA(keywords,filename):
    for i, keyword in enumerate(keywords): # 원하는 키워드 필터링
        if(i == 0):
            temp = df_qa.loc[df_qa['감정_소분류']==keyword]
            temp_short = temp.loc[temp['사람문장3'].isna()]
        else:
            temp = df_qa.loc[df_qa['감정_소분류']==keyword]
            temp_short = temp_short.append(temp.loc[temp['사람문장3'].isna()])
    
    temp_short.drop(['사람문장3', '시스템문장3'], axis=1, inplace=True) # 필요 없는 부분 drop
    df = temp_short.copy()
    df.drop(['감정_대분류', '감정_소분류','사람문장1','시스템문장2'], axis=1, inplace=True)
    df.columns = ['question', 'answer']

    for i, keyword in enumerate(keywords): # 원하는 키워드 필터링
        if(i == 0):
            temp = df_qa.loc[df_qa['감정_소분류']==keyword]
            temp_long = temp.loc[temp['사람문장3'].isna() == False]
        else:
            temp = df_qa.loc[df_qa['감정_소분류']==keyword]
            temp_long = temp_long.append(temp.loc[temp['사람문장3'].isna() == False])
        
    temp = temp_long.drop(['감정_대분류', '감정_소분류', '사람문장1','시스템문장3'], axis=1)
    temp1 = temp.drop(['시스템문장2', '사람문장3'], axis=1)
    temp2 = temp.drop(['시스템문장1', '사람문장2'], axis=1)
    temp1.columns = ['question', 'answer']
    temp2.columns = ['question', 'answer']

    df = df.append(temp1)
    df = df.append(temp2)

    df['question'] = df['question'].str.replace('.', ' .')
    df['question'] = df['question'].str.replace('!', ' !')
    df['question'] = df['question'].str.replace('?', ' ?')

    df['answer'] = df['answer'].str.replace('.', ' .')
    df['answer'] = df['answer'].str.replace('!', ' !')
    df['answer'] = df['answer'].str.replace('?', ' ?')

    df.reset_index(drop=True, inplace=True)

    df.to_json(path_or_buf=filename, force_ascii=False, orient = 'records', indent=4)
```

```python
def preprocessing_persona(target, filename):
    persona = df_persona[target].dropna()
    del persona[0]
    persona = pd.DataFrame(persona, columns=[target])
    persona.reset_index(drop=True, inplace=True)
    persona = persona.squeeze()
    data = {'question':['example'], 'answer':['example']}
    persona_df = pd.DataFrame(data)

    for i in range(0, len(persona)-1):
        if i%2 == 0:
            persona_df = persona_df.append({'question': persona[i], 'answer':persona[i+1]}, ignore_index=True)

    persona_df.drop(index=0, inplace=True)
    persona_df.reset_index(drop=True, inplace=True)

    persona_df['question'] = persona_df['question'].str.replace('.', ' .')
    persona_df['question'] = persona_df['question'].str.replace('!', ' !')
    persona_df['question'] = persona_df['question'].str.replace('?', ' ?')

    persona_df['answer'] = persona_df['answer'].str.replace('.', ' .')
    persona_df['answer'] = persona_df['answer'].str.replace('!', ' !')
    persona_df['answer'] = persona_df['answer'].str.replace('?', ' ?')

    value = persona_df

    persona_df.to_json(path_or_buf=filename, force_ascii=False, orient = 'records', indent=4)

    return value
```

```python
def preprocessing_all(personality, qa_file, persona_df, filename):
    other_answer = pd.read_json(qa_file)['answer']
    other_answer.dropna(inplace=True)
    other_answer.reset_index(inplace=True, drop=True)  

    datasets = []

    for i in range(len(persona_df)):
        utterance = {}
        candidates = list(other_answer[random.sample(range(len(other_answer)), 9)])
        candidates.append(persona_df['answer'][i])
        if i == 0:
            history = [persona_df['question'][i]]
        elif i < 4:
            history = list(persona_df.iloc[:i, :].values.flatten()) + [persona_df['question'][i]]
        else:
            history = history[2:] + [persona_df['answer'][i-1]] + [persona_df['question'][i]]
        history = list(history)

        utterance['candidates'] = candidates
        utterance['history'] = history

        if i % 7 == 0:
            utterances = []
        utterances.append(utterance)

        if i % 7 == 6:
            dataset = {}
            dataset['personality'] = personality
            dataset['utterances'] = utterances
            datasets.append(dataset)


    result = {
        'train' : datasets,
        'valid' : datasets
    }

    with open(f'{filename}', 'w') as f:
        json.dump(result, f, ensure_ascii=False, indent=4)
```
